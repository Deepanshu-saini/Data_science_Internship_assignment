```python
import pandas as pd
rf= pd.read_csv(r"/content/SampleQuestionFile_q.txt",delimiter="###0",header=None,on_bad_lines="skip")
rf.to_csv(r"/content/csvfile.csv")
col=["sno","question No."]
df=pd.read_csv(r"/content/csvfile.csv",names=col,skiprows=1)
df
```

```python
data=df['question No.']
quesno=[]
ques=[]
option1=[]
option2=[]
option3=[]
option4=[]
for x in range(len(df)):
  if(data[x][0].isdigit()):
    if(data[x][1].isdigit()):
      quesno.append(data[x][0:2])
    else:
      quesno.append(data[x][0:1])
    ques.append(data[x][3:])
  if(data[x]=="@@@"):
    continue
  if(data[x][0:3]=="(a)"):
    option1.append(data[x][3:])
  if(data[x][0:3]=="(b)"):
    option2.append(data[x][3:])
  if(data[x][0:3]=="(c)"):
    option3.append(data[x][3:])
  if(data[x][0:3]=="(d)"):
    option4.append(data[x][3:])
df.drop(["sno"],axis=1,inplace=True)
df.drop(["question No."],axis=1,inplace=True)
for x in range(30):
  option1.append(None)
  option2.append(None)
  option3.append(None)
  option4.append(None)
dn=df.head(30)
dn.to_csv(r"/content/newfile.csv")
dn["quesno"]=quesno
dn["ques"]=ques
dn["option1"]=option1
dn["option2"]=option2
dn["option3"]=option3
dn["option4"]=option4
dn
```

```python
rfs= pd.read_csv(r"/content/SampleSolutionFile_s.txt",delimiter="###0",header=None,on_bad_lines="skip")
rfs.to_csv(r"/content/solcsvfile.csv")
col1=["sno","sol."]
ds=pd.read_csv(r"/content/solcsvfile.csv",names=col1,skiprows=1)
ds
```

```python
ans=ds["sol."]
ans_key={}
ans_sol={}
i=0
s=[]
for y in range(len(ans)):
  if(ans[y][0].isdigit()):
    s=[]
    if(ans[y][1].isdigit()):
      ans_key[ans[y][0:2]]=ans[y][4:]
    else:
      ans_key[ans[y][0]]=ans[y][3:]
    i+=1
  else:
    s.append(ans[y])
    ans_sol[str(i)]=s
dn["key"]=dn["quesno"].map(ans_key)
dn["solution"]=dn["quesno"].map(ans_sol)
dn
```

```python
s1=dn["solution"]
for x in range(len(s1)):
  s1[x]="".join(s1[x])
dn
```
